```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

SHOW_BIG_ARRAY = False # показывать большие массивы если True
SHOW_PLOT_ARRAY = True # показывать большие массивы в виде графика

# Dataset - https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data
```

```python
def plot_big_array(data): # функция для более наглядного изображения массива
    plt.pcolor(data, cmap='hot')
    plt.colorbar()
    plt.show()
```

# Задание 1 и 2

```python
data = np.genfromtxt('iris.data', delimiter=',', dtype=np.float64) # извлекаю данные методом np
```

```python
print(type(data)) # убеждаюсь, что имею дело с нужными типами данных
print(type(data[2][2]))
shape = data.shape # убеждаюсь, что массив двумерный
print(shape)
```

```python
data = data[:, :4] # удаляю таргет по индексу
```

```python
if SHOW_PLOT_ARRAY:
    plot_big_array(data)
if SHOW_BIG_ARRAY:
    print(data)
```

# Задание 3

```python
print(data[:, 0]) # значения в первой колонке

value_0_150 = [0, shape[0]] # для построения графиков
```

```python
mean_first_column = data.mean(0)[0] # определяю среднее первой колонки
print(f'Mean={mean_first_column}') 

# стараюсь не копировать данные для сохранения памяти

plt.title('Mean')
plt.plot(data[:, 0], 'o') # показываю данные
plt.plot(value_0_150, [mean_first_column, mean_first_column], '-r') # показываю среднее на графике
plt.grid(which='major', linestyle='--', color='k', linewidth=0.7) 
plt.grid(which='minor', linestyle='-.', color='gray', linewidth=0.3)
plt.ylabel('Data')
plt.xlim(value_0_150)
plt.minorticks_on()
```

```python
median_first_column = np.median(data, axis=0)[0] # определяю медиану
print(f"Median={median_first_column}")

plt.title("Median")
plt.plot(np.sort(data[:, 0]), "o") # показываю отсортированные данные
plt.plot(value_0_150, [median_first_column, median_first_column], "-r") # показываю медиану
plt.grid(which="major", linestyle="--", color="k", linewidth=0.7) 
plt.grid(which="minor", linestyle="-.", color="gray", linewidth=0.3)
plt.ylabel("Data")
plt.xlim(value_0_150)
plt.minorticks_on()
```

```python
std_first_column = np.std(data, axis=0)[0] # определяю среднеквадратичное отклонение
print(f'Standard deviation={std_first_column}')

plt.title('Standard deviation')
plt.plot(data[:, 0], 'o') # показываю данные
plt.plot(value_0_150, [mean_first_column, mean_first_column], '-r') # показываю среднее
plt.plot(value_0_150, [mean_first_column - std_first_column, mean_first_column - std_first_column], '--r') # показываю среднеквадратичное отклонение (-)
plt.plot(value_0_150, [mean_first_column - std_first_column*2, mean_first_column - std_first_column*2], '--r', linewidth=0.5) # показываю среднеквадратичное отклонение (2-)
plt.plot(value_0_150, [mean_first_column + std_first_column, mean_first_column + std_first_column], '--r') # показываю среднеквадратичное отклонение (+)
plt.plot(value_0_150, [mean_first_column + std_first_column*2, mean_first_column + std_first_column*2], '--r', linewidth=0.7) # показываю среднеквадратичное отклонение (2+)
plt.grid(which='major', linestyle='--', color='k', linewidth=0.7) 
plt.grid(which='minor', linestyle='-.', color='gray', linewidth=0.3)
plt.ylabel('Data')
plt.xlim(value_0_150)
plt.minorticks_on()
```

# Задание 4 и 5

```python
nan_positions = np.random.choice(600, 20, replace=False) # replace - для того чтобы не было повторений
data = np.ravel(data) # разворачиваю массив
data[nan_positions] = np.nan # приравниваю
data = np.reshape(data, (150, 4)) # сворачиваю обратно
# может быть не самый лучший способ, но в теории значения мы не перезаписываем, а переназначаем ссылки к значениям
# минус этого способа в том, что, похоже, нельзя установить seed

print(f'Number of nan = {np.count_nonzero(np.isnan(data))}') # проверяю коли-во nan во всем массиве
print(f'Number of nan in fist column = {np.count_nonzero(np.isnan(data[:, 0]))}') # проверяю коли-во nan в первой колонке

rows, cols = np.where(np.isnan(data)) # нахожу координаты nan для всего массива
print(f'Array with rows indexes where is nan: {rows}')
print(f'Array with cols indexes where is nan: {cols}')

nan_first_column = np.where(np.isnan(data[:, 0])) # нахожу координаты nan для первой колонки
print(f'Array with indexes where is nan in first column: {nan_first_column}\n')

if SHOW_PLOT_ARRAY:
    plot_big_array(data)
if SHOW_BIG_ARRAY:
    print(data)
```

# Задание 13 и 14

```python
def func_13(column):
    mean = np.mean(column)
    column[column < mean] *= 2
    column[column >= mean] /= 4
    return column
```

```python
print(f'In: {data[:, 2]}')
print(f'\n Mean of column = {np.mean(data[:, 2])}\n')
print(f'Out: {func_13(data[:, 2].copy())}')

if SHOW_PLOT_ARRAY:
    plot_big_array(data)
if SHOW_BIG_ARRAY:
    print('\n')
    data[:, 2] = func_13(data[:, 2])
    print(data)
```
